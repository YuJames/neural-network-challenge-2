## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
x_cols = ["Age", "BusinessTravel", "DistanceFromHome", "Education", "EducationField", "HourlyRate", "JobLevel", "JobRole", "MaritalStatus", "NumCompaniesWorked"]

# Create X_df using your selected columns
X_df = attrition_df[x_cols]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder

# examine
# X_train.head()
# ordinal encoding
ordinal_columns = ["BusinessTravel"]
X_train[ordinal_columns].value_counts()
ordinal_encoder = OrdinalEncoder(categories=[["Non-Travel", "Travel_Rarely", "Travel_Frequently"]])
ordinal_encoder.fit(X_train[ordinal_columns])
X_train.loc[:, ordinal_columns] = ordinal_encoder.transform(X_train[ordinal_columns])
X_test.loc[:, ordinal_columns] = ordinal_encoder.transform(X_test[ordinal_columns])
# label encoding
label_columns = ["EducationField", "JobRole", "MaritalStatus"]
for col in label_columns:
    label_encoder = LabelEncoder()
    label_encoder.fit(X_train[col])
    X_train.loc[:, col] = label_encoder.transform(X_train[col])
    X_test.loc[:, col] = label_encoder.transform(X_test[col])

X_train.head()
```

```python
# Create a StandardScaler
std_scaler = StandardScaler()

# Fit the StandardScaler to the training data
std_scaler.fit(X_train)
# Scale the training and testing data
X_train_scaled = std_scaler.transform(X_train)
X_test_scaled = std_scaler.transform(X_test)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
onehot_encoder_department = OneHotEncoder()

# Fit the encoder to the training data
onehot_encoder_department.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_department_encoded = onehot_encoder_department.transform(y_train[['Department']])
y_test_department_encoded = onehot_encoder_department.transform(y_test[['Department']])
```

```python
# Create a OneHotEncoder for the Attrition column
onehot_encoder_attrition = OneHotEncoder()

# Fit the encoder to the training data
onehot_encoder_department.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attrition_encoded = onehot_encoder_department.transform(y_train[['Attrition']])
y_test_attrition_encoded = onehot_encoder_department.transform(y_test[['Attrition']])
```
